```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
#looking at the latest dates
df.tail()
```

```python
today=df[df['date']=='2020-07-17']
```

```python
today.head()
```

```python
max_confirmed_case=today.sort_values(by='confirmed',ascending=False)
```

```python
top_states_confirmed=max_confirmed_case[0:5]
```

```python
#making a bar plot for states with top confirmed cases
sns.set(rc={'figure.figsize':(10,7)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
```

```python
max_death_cases=today.sort_values(by='deaths',ascending=False)
max_death_cases
```

```python
top_state_death=max_death_cases[0:5]
top_state_death
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='state',y='deaths',data=top_state_death,hue='state')
plt.show()
```
